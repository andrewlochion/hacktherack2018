```python
import numpy as np
import pandas as pd
```

```python
df_posQty_posAmount = pd.read_csv('tran_posQty_posAmount.csv')
df_posQty_negAmount = pd.read_csv('tran_posQty_negAmount.csv')
df_posQty_zeroAmount = pd.read_csv('tran_posQty_zeroAmount.csv')
```

```python
b_hour = [5,6,7,8,9,10]
l_hour = [11,12,13,14]
t_hour = [15,16,17]
d_hour = [18,19,20,21,22]
s_hour = [0,1,2,3,4,23]
```

```python
df_posQty_posAmount_b = df_posQty_posAmount[df_posQty_posAmount['Hour'].isin(b_hour)]
df_posQty_posAmount_l = df_posQty_posAmount[df_posQty_posAmount['Hour'].isin(l_hour)]
df_posQty_posAmount_t = df_posQty_posAmount[df_posQty_posAmount['Hour'].isin(t_hour)]
df_posQty_posAmount_d = df_posQty_posAmount[df_posQty_posAmount['Hour'].isin(d_hour)]
df_posQty_posAmount_s = df_posQty_posAmount[df_posQty_posAmount['Hour'].isin(s_hour)]
```

```python
print(len(df_posQty_posAmount_b))
print(len(df_posQty_posAmount_l))
print(len(df_posQty_posAmount_t))
print(len(df_posQty_posAmount_d))
print(len(df_posQty_posAmount_s))
```

```python
df_posQty_posAmount_b.info()
```

```python
df_posQty_posAmount_b = df_posQty_posAmount_b[df_posQty_posAmount_b['Dept'].notnull()]
df_posQty_posAmount_l = df_posQty_posAmount_l[df_posQty_posAmount_l['Dept'].notnull()]
df_posQty_posAmount_t = df_posQty_posAmount_t[df_posQty_posAmount_t['Dept'].notnull()]
df_posQty_posAmount_d = df_posQty_posAmount_d[df_posQty_posAmount_d['Dept'].notnull()]
df_posQty_posAmount_s = df_posQty_posAmount_s[df_posQty_posAmount_s['Dept'].notnull()]
```

```python
print(len(df_posQty_posAmount_b))
print(len(df_posQty_posAmount_l))
print(len(df_posQty_posAmount_t))
print(len(df_posQty_posAmount_d))
print(len(df_posQty_posAmount_s))
```

```python
b_dept_pct = df_posQty_posAmount_b[df_posQty_posAmount_b['Dept']!='Convenience Services']['Dept'].value_counts(normalize=True)
b_dept_pct
```

```python
l_dept_pct = df_posQty_posAmount_l[df_posQty_posAmount_l['Dept']!='Convenience Services']['Dept'].value_counts(normalize=True)
l_dept_pct
```

```python
t_dept_pct = df_posQty_posAmount_t[df_posQty_posAmount_t['Dept']!='Convenience Services']['Dept'].value_counts(normalize=True)
t_dept_pct
```

```python
d_dept_pct = df_posQty_posAmount_d[df_posQty_posAmount_d['Dept']!='Convenience Services']['Dept'].value_counts(normalize=True)
d_dept_pct
```

```python
s_dept_pct = df_posQty_posAmount_s[df_posQty_posAmount_s['Dept']!='Convenience Services']['Dept'].value_counts(normalize=True)
s_dept_pct
```

```python
b_dept_pct.to_csv('breakfast_dept.csv')
l_dept_pct.to_csv('lunch_dept.csv')
t_dept_pct.to_csv('tea_dept.csv')
d_dept_pct.to_csv('dinner_dept.csv')
s_dept_pct.to_csv('supper_dept.csv')
```

```python
from matplotlib import pyplot as plt

all_dept_pct = pd.concat([b_dept_pct,l_dept_pct,t_dept_pct,d_dept_pct,s_dept_pct],keys=['Breakfast','Lunch','Tea','Dinner','Supper'],axis=1,sort=True)
all_dept_pct = all_dept_pct.sort_values(by=['Tea'], ascending=False)
all_dept_pct
```

```python
others = pd.DataFrame(all_dept_pct.iloc[7:].sum(), columns=['Others'])
all_dept_pct = all_dept_pct.drop(all_dept_pct.index[7:])
all_dept_pct = pd.concat([all_dept_pct, others.T])
```

```python
all_dept_pct
```

```python
all_dept_pct['Breakfast'].plot(kind='pie', figsize=(8, 8))
plt.title('Transactions - Breakfast', fontsize=20)
plt.ylabel('')
plt.savefig('all_breakfast.jpg')
```
