```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_posQty_posAmount = pd.read_csv('tran_posQty_posAmount.csv')
```

```python
df_posQty_posAmount.info()
```

```python
df_posQty_posAmount.datetime = pd.to_datetime(df_posQty_posAmount.datetime)
```

```python
df_posQty_posAmount.info()
```

```python
df_posQty_posAmount.head()
```

```python
df_posQty_posAmount['datetime'].apply(lambda x: x.month)
```

```python
df_posQty_posAmount['month'] = df_posQty_posAmount['datetime'].apply(lambda x: x.month)
```

```python
df_posQty_posAmount['day'] = df_posQty_posAmount['datetime'].apply(lambda x: x.day)
```

```python
df_posQty_posAmount.head()
```

```python
df_posQty_posAmount = df_posQty_posAmount[df_posQty_posAmount.month==4]
```

```python
df_posQty_posAmount['weekday'] = df_posQty_posAmount['datetime'].apply(lambda x: x.weekday())
```

```python
df_posQty_posAmount['weekday'].value_counts()
```

```python
df_posQty_posAmount['holiday'] = np.where(np.logical_or(df_posQty_posAmount['weekday']==5,df_posQty_posAmount['weekday']==6), 1, 0) #or (df_posQty_posAmount['weekday']==6) or (df_posQty_posAmount['day']==2) or (df_posQty_posAmount['day']==5), 1, 0)
```

```python
df_posQty_posAmount.loc[df_posQty_posAmount['day']==2,'holiday']=1
df_posQty_posAmount.loc[df_posQty_posAmount['day']==5,'holiday']=1
```

```python
df_posQty_posAmount.head()
```

```python
df_posQty_posAmount[df_posQty_posAmount['day']==5]
```

```python
df_holiday = df_posQty_posAmount[df_posQty_posAmount['holiday']==1]
df_workingday = df_posQty_posAmount[df_posQty_posAmount['holiday']==0]
```

```python
print(len(df_holiday))
print(len(df_workingday))
```

```python
b_hour = [5,6,7,8,9,10]
l_hour = [11,12,13,14]
t_hour = [15,16,17]
d_hour = [18,19,20,21,22]
s_hour = [0,1,2,3,4,23]
```

```python
df_holiday_b = df_holiday[df_holiday['Hour'].isin(b_hour)]
df_holiday_l = df_holiday[df_holiday['Hour'].isin(l_hour)]
df_holiday_t = df_holiday[df_holiday['Hour'].isin(t_hour)]
df_holiday_d = df_holiday[df_holiday['Hour'].isin(d_hour)]
df_holiday_s = df_holiday[df_holiday['Hour'].isin(s_hour)]
```

```python
df_workingday_b = df_workingday[df_workingday['Hour'].isin(b_hour)]
df_workingday_l = df_workingday[df_workingday['Hour'].isin(l_hour)]
df_workingday_t = df_workingday[df_workingday['Hour'].isin(t_hour)]
df_workingday_d = df_workingday[df_workingday['Hour'].isin(d_hour)]
df_workingday_s = df_workingday[df_workingday['Hour'].isin(s_hour)]
```

```python
print(len(df_holiday_b))
print(len(df_holiday_l))
print(len(df_holiday_t))
print(len(df_holiday_d))
print(len(df_holiday_s))
```

```python
print(len(df_workingday_b))
print(len(df_workingday_l))
print(len(df_workingday_t))
print(len(df_workingday_d))
print(len(df_workingday_s))
```

```python
holiday_b_dept_pct = df_holiday_b[df_holiday_b['Dept']!='Convenience Services']['Dept'].value_counts(normalize=True)
holiday_b_dept_pct
```

```python
holiday_l_dept_pct = df_holiday_l[df_holiday_l['Dept']!='Convenience Services']['Dept'].value_counts(normalize=True)
holiday_l_dept_pct
```

```python
holiday_t_dept_pct = df_holiday_t[df_holiday_t['Dept']!='Convenience Services']['Dept'].value_counts(normalize=True)
holiday_t_dept_pct
```

```python
holiday_d_dept_pct = df_holiday_d[df_holiday_d['Dept']!='Convenience Services']['Dept'].value_counts(normalize=True)
holiday_d_dept_pct
```

```python
holiday_s_dept_pct = df_holiday_s[df_holiday_s['Dept']!='Convenience Services']['Dept'].value_counts(normalize=True)
holiday_s_dept_pct
```

```python
workingday_b_dept_pct = df_workingday_b[df_workingday_b['Dept']!='Convenience Services']['Dept'].value_counts(normalize=True)
workingday_b_dept_pct
```

```python
workingday_l_dept_pct = df_workingday_l[df_workingday_l['Dept']!='Convenience Services']['Dept'].value_counts(normalize=True)
workingday_l_dept_pct
```

```python
holiday_all_dept_pct = pd.concat([holiday_b_dept_pct,holiday_l_dept_pct,holiday_t_dept_pct,holiday_d_dept_pct,holiday_s_dept_pct],keys=['Breakfast','Lunch','Tea','Dinner','Supper'],axis=1,sort=True)
holiday_all_dept_pct = holiday_all_dept_pct.sort_values(by=['Tea'], ascending=False)
holiday_all_dept_pct
```

```python
workingday_all_dept_pct = pd.concat([workingday_b_dept_pct,workingday_l_dept_pct,workingday_t_dept_pct,workingday_d_dept_pct,workingday_s_dept_pct],keys=['Breakfast','Lunch','Tea','Dinner','Supper'],axis=1,sort=True)
workingday_all_dept_pct = workingday_all_dept_pct.sort_values(by=['Tea'], ascending=False)
workingday_all_dept_pct
```

```python
others = pd.DataFrame(workingday_all_dept_pct.iloc[7:].sum(), columns=['Others'])
workingday_all_dept_pct = workingday_all_dept_pct.drop(workingday_all_dept_pct.index[7:])
workingday_all_dept_pct = pd.concat([workingday_all_dept_pct, others.T])
workingday_all_dept_pct
```

```python
holiday_all_dept_total = pd.concat([holiday_b_dept_total,holiday_l_dept_total,holiday_t_dept_total,holiday_d_dept_total,holiday_s_dept_total],keys=['Breakfast','Lunch','Tea','Dinner','Supper'],axis=1,sort=True)
holiday_all_dept_total['Total'] = holiday_all_dept_total.sum(axis=1)
holiday_all_dept_total = holiday_all_dept_total.sort_values(by=['Total'], ascending=False)
holiday_all_dept_total
```

```python
workingday_all_dept_total = pd.concat([workingday_b_dept_total,workingday_l_dept_total,workingday_t_dept_total,workingday_d_dept_total,workingday_s_dept_total],keys=['Breakfast','Lunch','Tea','Dinner','Supper'],axis=1,sort=True)
workingday_all_dept_total['Total'] = workingday_all_dept_total.sum(axis=1)
workingday_all_dept_total = workingday_all_dept_total.sort_values(by=['Total'], ascending=False)
workingday_all_dept_total
```

```python
holiday_all_dept_dailyavg = holiday_all_dept_total.divide(11)
holiday_all_dept_dailyavg
```
